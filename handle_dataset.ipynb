```python
import csv
from ast import literal_eval

path = "Grab SG Restaurants.csv"
modified_path = "Grab SG Restaurants Modified.csv"

# headers = ["id_source", "name", "address", "country", "cuisine", "currency", "delivery_cost", "lat", "lon", "opening_hours"]
with open(path, 'r') as file:
    new_rows = []
    csv_reader = csv.reader(file)

    headers = next(csv_reader)
    # new headers
    new_headers = headers + ["halal", "beverage", "soup", "seafood", "healthy"]
    for row in csv_reader:
        # iterate through the cuisine section to pick out new columns to add
        isHalal = False
        isBeverage = False
        isSoup = False
        isSeafood = False
        isHealthy = False
        cuisine_tags = row[4]
        # cuisine_tags = literal_eval(cuisine_tags)
        cuisine_tags = cuisine_tags.strip("[]").replace('"', "").split(",")
        cuisine_tags = [tag.strip() for tag in cuisine_tags]

        # transform list to become lowercase
        cuisine_tags_lower = [tag.lower() for tag in cuisine_tags]
        # print(cuisine_tags_lower)

        if "halal" in cuisine_tags_lower:
            isHalal = True
        if "beverages" in cuisine_tags_lower:
            isBeverage = True
        if "soup" in cuisine_tags_lower:
            isSoup = True
        if "seafood" in cuisine_tags_lower:
            isSeafood = True
        if "healthy" in cuisine_tags_lower:
            isHealthy = True

        new_row = row + [isHalal, isBeverage, isSoup, isSeafood, isHealthy]
        new_rows.append(new_row)

with open(modified_path, "w", newline="") as file:
    csv_writer = csv.writer(file)
    csv_writer.writerow(new_headers)
    csv_writer.writerows(new_rows)

```
