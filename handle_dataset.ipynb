```python
import csv

path = "Grab SG Restaurants.csv"
modified_path = "Grab SG Restaurants Modified.csv"

countries_set = {"japanese", "korean", "thai", "western", "american", "vietnamese", "chinese", "italian", "asian", "malaysian", "indonesian", "indian"}

def getCountries(cuisine_tags_lower):
    # print("Is 'japanese' in countries?", 'japanese' in countries)
    res = []
    for tag in cuisine_tags_lower:
        tag_stripped = tag.strip()
        if tag_stripped in countries_set:
            res.append(tag_stripped)
    if len(res) == 0:
        return None
    return res
'''
Code to append new rows to the dataset 
'''
with open(path, 'r') as file:
    new_rows = []
    csv_reader = csv.reader(file)

    headers = next(csv_reader)
    # new headers
    new_headers = headers + ["halal", "beverage", "soup", "seafood", "healthy", "fast food", "local", "countries"]
    for row in csv_reader:
        # iterate through the cuisine section to identify features and append new rows
        isHalal = False
        isBeverage = False
        isSoup = False
        isSeafood = False
        isHealthy = False
        isFastFood = False
        isLocal = False
        cuisine_tags = row[4]
        cuisine_tags = cuisine_tags.strip("[]").replace('"', "").split(",")
        cuisine_tags = [tag.strip() for tag in cuisine_tags]

        # transform list to become lowercase
        cuisine_tags_lower = [tag.lower() for tag in cuisine_tags]
        # print(cuisine_tags_lower)

        if "halal" or "muslim" in cuisine_tags_lower:
            isHalal = True
        if "beverages" in cuisine_tags_lower:
            isBeverage = True
        if "soup" in cuisine_tags_lower:
            isSoup = True
        if "seafood" in cuisine_tags_lower:
            isSeafood = True
        if "healthy" in cuisine_tags_lower:
            isHealthy = True
        if "fast food" in cuisine_tags_lower:
            isFastFood = True
        if "local" in cuisine_tags_lower or "singaporean" in cuisine_tags_lower:
            isLocal = True

        countries = getCountries(cuisine_tags_lower)
        if not countries:
            formatted_countries = '["]'
        else:
            formatted_countries = '["' + '", "'.join(countries) + '"]'
        new_row = row + [isHalal, isBeverage, isSoup, isSeafood, isHealthy, isFastFood, isLocal, formatted_countries]
        print("isLocal", isLocal)
        print(new_row)
        new_rows.append(new_row)

with open(modified_path, "w", newline="") as file:
    csv_writer = csv.writer(file)
    csv_writer.writerow(new_headers)
    csv_writer.writerows(new_rows)

```
